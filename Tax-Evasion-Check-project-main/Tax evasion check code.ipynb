```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt 
df= pd.read_excel('tax evasion Project data.xlsx')
```

```python
df.head()
```

```python
df.columns
```

```python
df['SP_through_outlets']=df['Product sales through sales outlets'] / df['The Quantity Sold']
```

```python
df.head()
```

```python
df.rename(columns={'test< ':'test1','test>':'test2','test=':'test3', 'The price of linking with the Zakat Authority system':'TPWZAS', 'Selling price through outlets':'SPTO'}, inplace=True)
df.head()
```

```python
df['test']='سماح'
df.head(10)
```

```python
for index, row in df.iterrows():
    if row['TPWZAS']< row['SPTO']:
        df.at[index,'test']='متهرب'
    elif row['TPWZAS']== row['SPTO']:
        df.at[index,'test']= 'غير متهرب'
        row['test']= 'غير متهرب'
df.head(10)
```

```python
df.info()
```

```python
df.drop(['SPTO','test1','test2','test3'],axis=1, inplace=True)
df.head()
```

```python
feature_columns=['TPWZAS','SP_through_outlets']
X=df[feature_columns]
X
```

```python
y=df['test']
y
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
y=lb.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=4,shuffle=True)
```

```python
X_train
```

```python
X_test
```

```python
### SVM (Support Vector Machine)
from sklearn.svm import SVC
def svm_model(X_train,y_train,X_test):
    svm = SVC(kernel='linear', C=1,random_state=0)
    svm.fit(X_train,y_train)
    y_pred=svm.predict(X_test)
    return y_pred
```

```python
y_pred=svm_model(X_train,y_train,X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_sklearn = accuracy_score(y_test, y_pred)*100
print('Model Accuracy:',accuracy_sklearn)
```

```python
from sklearn.linear_model import LogisticRegression
def LR_model(X_train,y_train,X_test):
    LR = LogisticRegression()
    LR.fit(X_train,y_train)
    y_pred=LR.predict(X_test)
    return y_pred
```

```python
y_pred=LR_model(X_train,y_train,X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_sklearn = accuracy_score(y_test, y_pred)*100
print('Model Accuracy:',accuracy_sklearn)
```

```python
from sklearn.neighbors import KNeighborsClassifier
def knn_model(X_train,y_train,X_test,k):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    return y_pred
```

```python
y_pred=knn_model(X_train,y_train,X_test,k=1)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_sklearn = accuracy_score(y_test, y_pred)*100
print('Model Accuracy:',accuracy_sklearn)
```

```python
k_value=range(1,30)
accuracy=[]
for k in k_value:
    y_predict=knn_model(X_train,y_train,X_test,k)
    accur=accuracy_score(y_test,y_predict)
    accuracy.append(accur)
```

```python
plt.xlabel('k')
plt.ylabel('accuracy')
plt.plot(k_value,accuracy,c='g')
plt.show()
```
